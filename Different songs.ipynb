```python
import librosa
import pandas as pd
import plotly.plotly as py
import IPython.display as ipd
import numpy as np

y1, sr1 = librosa.load("C://Users//Chinmay//Downloads//02 Bring It On Full Song.mp3")
y2, sr2 = librosa.load("C://Users//Chinmay//Downloads//13 Ye Go Ye Ye Maina(VipMarathi.in).mp3")
y3, sr3 = librosa.load("C://Users//Chinmay//Downloads//01 Dolby Walya Full Song.mp3")

yz_harmonic, yz_percussive1 = librosa.effects.hpss(y1)
yz_harmonic, yz_percussive2 = librosa.effects.hpss(y2)
yz_harmonic, yz_percussive3 = librosa.effects.hpss(y3)

tempo1, beats1 = librosa.beat.beat_track(y=yz_percussive1, sr=sr1)
tempo3, beats3 = librosa.beat.beat_track(y=yz_percussive2, sr=sr3)
sr3= (tempo1/tempo3)*sr1

for i in range(3) :
    a_harmonic1, yz_percussive1 = librosa.effects.hpss(yz_percussive1)
    a_harmonic2, yz_percussive2 = librosa.effects.hpss(yz_percussive2)
    a_harmonic3, yz_percussive3 = librosa.effects.hpss(yz_percussive3)
    
```

```python
import matplotlib.pyplot as plt
import csv   
odf = librosa.onset.onset_strength(y=yz_percussive1, sr=sr1, hop_length=512)
ac1 = librosa.autocorrelate(odf, max_size= 100)
with open('C://Users//Chinmay//Downloads//beat_dataset.csv', 'w') as writeFile:
    writer = csv.writer(writeFile)
    writer.writerow(ac1)

plt.plot(ac1)
```

```python
y4, sr4 = librosa.load("C://Users//Chinmay//Downloads//Zing Zing Zingat - Ajay Atul (1) extract .mp3")
yz_harmonic, yz_percussive4 = librosa.effects.hpss(y4)

for i in range(3) :
    a_harmonic1, yz_percussive4 = librosa.effects.hpss(yz_percussive4)

tempo4, beats4 = librosa.beat.beat_track(y=yz_percussive4, sr=sr4)
sr4= (tempo1/tempo4)*sr1

y_new=librosa.effects.time_stretch(y4, tempo1/tempo4)
                             
odf = librosa.onset.onset_strength(y=y_new, sr=sr4, hop_length=512)
ac1 = librosa.autocorrelate(odf, max_size= 100)

with open('C://Users//Chinmay//Downloads//beat_dataset.csv', 'a') as f:
    writer = csv.writer(f)
    writer.writerow(ac1)

plt.plot(ac1)
```
